#### dataset_sophia2 
Las noticias descargadas desde caleuche se almacenan en una estructura de datos llamada [DataFrame](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html) de la libreria pandas en donde cada fila corresponde a una noticia.

Este dataset cuenta con las siguientes columnas para cada noticia:

* id_news: corresponde a un id unico interno asignado por la base de datos a la noticia
* country: pais al que pertenece el medio donde fue publicada 
* media_outlet: nombre del medio de prensa donde fue publicada
* url: link desde donde fue extraida
* title: titulo 
* text: cuerpo 
* date: fecha de publicacion
* search: palabras clave

# 2) TOOLS

```python
from pysophia2 import tools
```

#### frequency_struct
Para las funciones que vienen a continuacion es necesario pasar como argumento una "word_disgtribution". Esta estructura es creada a partir de un [dataset](#dataset) cualquiera mediante la funcion [`ngram.calculateDist()`](#ngram.calculateDist(dataset,group_by)) y se hace con el objetivo de optimizar los tiempos de ejecucion en funciones que requieren previamente un paso en el que se obtiene un diccionario de frecuencias.


la estructura de frecuencias es un diccionario de python que contiene los siguientes campos:

* total_elements: cantidad totales de palabras en las noticias 
* dictionary: diccionario de frecuencias en donde aparece la cantidad de apariciones para cada palabra


#### ngram.calculateDist(dataset,group_by)
<div style='text-align: right'><a href="#INDEX">back to index </a></div>

```python
%%time
word_distribution = tools.ngram.calculateDist(dataset=df, group_by="month")#puede ser day|month|year
word_distribution
```

#### ngram.view(word_distribution,keywords,annotate)
<div style='text-align: right'><a href="#INDEX">back to index </a></div>

```python
%%time
keywords=["cambio clim√°tico", "diversidad", "ecolog√≠a", "biodiversidad"]
tools.ngram.view(word_distribution=word_distribution,words=keywords, annotate=True, normalize=True)
```

#### ngram.freq(word_distribution,top_n)
<div style='text-align: right'><a href="#INDEX">back to index </a></div>

```python
tools.ngram.freq(word_distribution=word_distribution,top_n=50)
```

#### ngram.freq(word_distribution,words)
<div style='text-align: right'><a href="#INDEX">back to index </a></div>

```python
array=tools.ngram.freq(word_distribution,words=["biodiversidad", "clima"])
```
