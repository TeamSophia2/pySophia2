#### dataset_sophia2 
Las noticias descargadas desde caleuche se almacenan en una estructura de datos llamada [DataFrame](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html) de la libreria pandas en donde cada fila corresponde a una noticia.

Este dataset cuenta con las siguientes columnas para cada noticia:

* id: corresponde a un id unico interno asignado por la base de datos a la noticia
* country: pais al que pertenece el medio donde fue publicada 
* media_outlet: nombre del medio de prensa donde fue publicada
* url: link desde donde fue extraida
* title: titulo 
* text: cuerpo 
* date: fecha de publicacion
* year: año de publicacion
* journalist: autor de la noticia
* month: mes de publicacion

# 2) TOOLS

```python
from sophia2 import tools
```

#### frequency_struct
Para muchas de las funciones que vienen a continuacion es necesario pasar como argumento una "frecuency_struct". Esta estructura es creada a partir de un [dataset](#dataset) cualquiera mediante la funcion [`get_frecuency()`](#get_frecuency(dataset)) y se hace con el objetivo de optimizar los tiempos de ejecucion en funciones que requieren previamente un paso en el que se obtiene un diccionario de frecuencias.


la estructura de frecuencias es un diccionario de python que contiene los siguientes campos:

* total_elements: cantidad totales de palabras en las noticias 
* dictionary: diccionario de frecuencias en donde aparece la cantidad de apariciones para cada palabra


### get_frecuency([dataset_sophia2](#dataset_sophia2))
<div style='text-align: right'><a href="#INDEX">back to index </a></div>

```python
%%time
dataset = db.dataset_by_country('chile','2020-03-28','2020-04-02')
```

```python
%%time
tools.get_frecuency(dataset)
```

### get_n_most_freq([freq_struct](#frequency_struct),n,graph)

retorna una lista de n tuplas en donde cada tupla esta compuesta por la palabra y la cantidad de apariciones en los textos terminos mas frecuentes de una estructura de frecuencias, opcionalmente puedes pasar el parametro graph=true para obtener un grafico en terminos de la frecuencia para cada uno de estos n terminos
<div style='text-align: right'><a href="#INDEX">back to index </a></div>

```python
%%time
frequency_struct = tools.get_frecuency(dataset)
```

```python
%%time 
tools.get_n_most_freq(frequency_struct,5, True)
```

### graph_term_frecuency_bar([freq_struct](#frequency_struct),[word])
<div style='text-align: right'><a href="#INDEX">back to index </a></div>

```python
%%time
tools.graph_term_frecuency_bar(frequency_struct,["personas", "sebastian", "coronavirus", "covid", "piñera", "medidas"])
```

### build_sophia2_ngram_struct([dataset_sophia2](#dataset_sophia2),granularity) 
<div style='text-align: right'><a href="#INDEX">back to index </a></div>

Permite generar una estructura sophia2_ngram_struct compuesta por un diccionario de python con los dias, meses o años y una estructura de frecuencias asociada a este periodo, para seleccionar los periodos existen 3 opciones: 
* granularity = "day" | "month" | "year"

#consultar con matthiew si abstraemos al usuario final de esto

```python
%%time
sophia2_dataset = db.dataset_by_country('chile','2020-12-28','2021-01-02')
```

```python
sophia2_ngram_struct = tools.build_sophia2_ngram_struct(sophia2_dataset, granularity="day")
sophia2_ngram_struct
```

### sophia_ngram_view(sophia2_ngram_struct,[palabra],annotations)
<div style='text-align: right'><a href="#INDEX">back to index </a></div>

Permite generar una visualizacion de la evolucion temporal de un grupo de palabras basada en la herramienta Google NGramViewer. Es necesario pasar como parametro una **sophia2_ngram_struct** y una lista de palabras a visualizar en la grafica

```python
palabras = ["salud","personas", "caso", "millones"]
tools.sophia_ngram_view(sophia2_ngram_struct, palabras,True)
```
